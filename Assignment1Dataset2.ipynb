# Part 1

```python
import pandas as pd

df = pd.read_csv('historical-weather.csv')
df
```

## Cleaning DataSet by dropping null values

```python
df = df.dropna()
df
```

## Preprocess data to deal with outliers 

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df.iloc[:,2:])
scaled = scaler.transform(df.iloc[:,2:])
scaled
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df.air_pressure, df.avg_wind_direction, df.rain_duration)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

# Part 3

## My Kmeans

```python
import matplotlib.pyplot as pl
import pandas as pd
from KMeans import KMeans as my_kmeans

km = my_kmeans()
km.fit(scaled)
km.labels_
```

```python
len(scaled[0])
```

### Scatter Plot

```python

d = pd.DataFrame(scaled, columns = ["column" for i in range(11)])
d['Cluster'] = km.labels_

fig = pl.figure()
ax = fig.add_subplot(projection='3d')

colors = {0: 'red', 1: 'green', 2: 'blue'}
ax.scatter(df.air_pressure, df.avg_wind_direction, df.rain_duration, c=d['Cluster'].map(colors))

pl.show()
```

## Sklearn's Kmeans

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(scaled)
kmeans.labels_
```

### Scatter Plot

```python

d = pd.DataFrame(scaled, columns = ["column" for i in range(11)])
d['Cluster'] = kmeans.labels_

fig = pl.figure()
ax = fig.add_subplot(projection='3d')

colors = {0: 'red', 1: 'green', 2: 'blue'}
ax.scatter(df.air_pressure, df.avg_wind_direction, df.rain_duration, c=d['Cluster'].map(colors))

pl.show()
```

## Metrics

```python
from sklearn.metrics import v_measure_score
v_measure_score(kmeans.labels_, km.labels_)
```
